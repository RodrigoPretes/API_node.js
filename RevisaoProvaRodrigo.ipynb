```python

import tensorflow as tf
print(tf.__version__)

import keras
import pandas as pd
import numpy as np

# import warnings
# warnings.filterwarnings('ignore')



# Entra com os dados
X_treino = np.array(
    [
        [350,6], [200,3], [180,4], [400,7], [380,8],
        [410,7], [220,2], [190,3]
    ]
)
Y_treino = np.array( [ #-1 laranja / # 1 limao
         [-1],[1],[1],[-1],[-1],
         [-1],[1],[1] ] )
X_teste = np.array(
    [
        [390,8], [160,3]
    ]
)
Y_teste = np.array( [[1],[-1]] )




# Monta a rede
from keras.models import Sequential
from keras.layers import Dense

modelo = Sequential() # Inicializa a Rede Neural Artificial

modelo.add(Dense(units = 1, input_dim = X_treino.shape[1]))



# Mostra a rede
modelo.summary()



# Treina a rede
modelo.compile(loss='mae', optimizer = 'adam', metrics=['mae'])


resultado = modelo.fit(X_treino, Y_treino, batch_size = 13, epochs = 2000, validation_data=(X_teste, Y_teste))



# Mostra o Teste
Y_predito = modelo.predict(X_teste)

Y_predito





# Teste da Rede
Y_predito = modelo.predict(X_teste) # Faz o arrendodamento
Y_predito = [ 1 if y>=0.0 else -1 for y in Y_predito ] #positivo +1 / negativo -1
print("Valores Preditos:",Y_predito)

total = 0
correto = 0
errado = 0
for i in Y_predito:
  total=total+1
  if(Y_teste[i,0] == Y_predito[i]):
    correto=correto+1
  else:
    errado=errado+1

print("Total " + str(total))
print("Correto " + str(correto))
print("Errado " + str(errado))



# Mostra graficos
import matplotlib.pyplot as plt

plt.plot(resultado.history['loss'])
plt.plot(resultado.history['val_loss'])
plt.title('HistÃ³rico de Treinamento')
plt.ylabel('FunÃ§Ã£o de custo')
plt.xlabel('Ãpocas de treinamento')
plt.legend(['Erro treino', 'Erro teste'])
plt.show()





# Mostra Pesos
for layerNum, layer in enumerate(modelo.layers):
    weights = layer.get_weights()[0]
    biases = layer.get_weights()[1]

    for toNeuronNum, bias in enumerate(biases):
        print(f'{layerNum}B -> L{layerNum+1}N{toNeuronNum}: {bias}')

    for fromNeuronNum, wgt in enumerate(weights):
        for toNeuronNum, wgt2 in enumerate(wgt):
            print(f'L{layerNum}N{fromNeuronNum} \
                  -> L{layerNum+1}N{toNeuronNum} = {wgt2}')
```

```python
# Criando as variÃ¡veis do problema
valor1 = ctrl.Antecedent(np.arange(0,1,0.1),'valor1')
valor2 = ctrl.Antecedent(np.arange(0,0.9,0.1),'valor2')
And = ctrl.Consequent(np.arange(0,1,0.1),'And')
Or = ctrl.Consequent(np.arange(0,1,0.1),'Or')
```

```python
# Criando as funÃ§Ãµes de pertinÃªncia para o valor 1
valor1['0']= fuzz.trapmf(valor1.universe,[0.0,0.0,0.4,0.5])
valor1['1']= fuzz.trapmf(valor1.universe,[0.4,0.5,1.0,1.0])
```

```python
valor1.view()
```

```python
# Criando as funÃ§Ãµes de pertinÃªncia para o valor 2

valor2['0']= fuzz.trapmf(valor2.universe,[0.0,0.0,0.3,0.5])
valor2['1']= fuzz.trapmf(valor2.universe,[0.3,0.5,0.9,0.9])

valor2.view()
```

```python
And['0']= fuzz.trapmf(And.universe,[0.0,0.0,0.7,0.7])
And['1']= fuzz.trapmf(And.universe,[0.7,0.7,1.0,1.0])

And.view()
```

```python
Or['0']= fuzz.trapmf(Or.universe,[0.0,0.0,0.7,0.7])
Or['1']= fuzz.trapmf(Or.universe,[0.7,0.7,1.0,1.0])

Or.view()
```

```python
# Base de Conhecimento/Regras
rule1 = ctrl.Rule(valor1['0'],And['0'])
rule2 = ctrl.Rule(valor2['0'],And['0'])
rule3 = ctrl.Rule(valor1['1'] & valor2['1'],And['1'])
rule4 = ctrl.Rule(valor1['1'] | valor2['1'],Or['1'])
rule5 = ctrl.Rule(valor1['1'],Or['1'])
rule6 = ctrl.Rule(valor2['1'],Or['1'])
```

```python
# Sistema Fuzzy e SimulaÃ§Ã£o
validade_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6])
validade_simulador = ctrl.ControlSystemSimulation(validade_ctrl)
```

```python
# Entranda do dinheiro
while True:
   val1 = float(input('Digite valor 1): '))
   if (val1 < 0.0 or val1 > 1.0):
      print('O valor deve estar no intervalo [0, 1]')
      continue
   validade_simulador.input['valor1'] = val1
   break

# Entrada do pessoal
while True:
    val2 = float(input('Digite valor 2): '))
    if (val2 < 0.0 or val2 > 1.0):
       print('o pessoal deve estar no intervalo [0, 1]')
       continue
    validade_simulador.input['valor2'] = val2
    break

validade_simulador.compute()
print('And Ã© %.1F' %validade_simulador.output['And'])
print('Or Ã© %.1F' %validade_simulador.output['Or'])
```

```python
!pip install geneticalgorithm
import numpy as np
import math
from geneticalgorithm import geneticalgorithm as ga

b = np.array([
    [0, 1, 7, 5, 3],
    [1, 0, 6, 2, 5],
    [7, 6, 0, 4, 5],
    [4, 2, 4, 0, 1],
    [3, 5, 5, 1, 0]
])

inicio=0
fim=0

def f(X):
    dim=len(X)
    OF=0
    OF+=b[inicio][math.trunc(X[0])]
    for i in range(1,dim):
      OF+=b[math.trunc(X[i-1])][math.trunc(X[i])]

    for i in range(0,dim-1):
     for j in range(i+1,dim):
       if math.trunc(X[i])==math.trunc(X[j]):
         OF+=999

    OF+=b[math.trunc(X[dim-1])][fim]
    return OF

varbound=np.array([[0,len(b)-1]]*len(b))

algorithm_param = {'max_num_iteration': 3000,\
                   'population_size':100,\
                   'mutation_probability':0.1,\
                   'elit_ratio': 0.01,\
                   'crossover_probability': 0.5,\
                   'parents_portion': 0.3,\
                   'crossover_type':'uniform',\
                   'max_iteration_without_improv':None}

model = ga(function=f, dimension=len(b), variable_type='int', variable_boundaries=varbound, algorithm_parameters=algorithm_param)
model.run()


cidade = np.array(["A", "B", "C", "D", "E"])
print(cidade[inicio])
for i in model.best_variable:
    print(cidade[math.trunc(i)])

print(cidade[fim])
```
